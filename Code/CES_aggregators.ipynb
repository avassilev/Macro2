```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
theta = 4
assert theta > 1
J = np.linspace(0,1,num=100)
C = 5

def pfun(x):
    return np.exp(x)

p = pfun(J)

def pintfun(x):
    return pfun(x)**(1-theta)
Pint = pintfun(J)
P,_ = quad(pintfun,0,1)
P = P**(1/(1-theta))

def cfun(x):
    return (pfun(x)/P)**(-theta)*C

c = cfun(J)


plt.figure()
plt.plot(J,p,label='$p(j)$')
plt.plot(J,Pint,label='$p(j)^{1-\\theta}$')
plt.legend()
plt.show()

plt.figure()
plt.plot(J,p,label='$p(j)$')
plt.plot(J,c,label='$c(j)$', color='r')
plt.legend()
plt.show()
```

```python
theta = 2
assert theta > 1
J = np.linspace(0,1,num=100)
C = 5

def pfun(x):
    return x*(1-x)+1

p = pfun(J)

def pintfun(x):
    return pfun(x)**(1-theta)
Pint = pintfun(J)
P,_ = quad(pintfun,0,1)
P = P**(1/(1-theta))

def cfun(x):
    return (pfun(x)/P)**(-theta)*C

c = cfun(J)


plt.figure()
plt.plot(J,p,label='$p(j)$')
plt.plot(J,Pint,label='$p(j)^{1-\\theta}$')
plt.legend()
plt.show()

plt.figure()
plt.plot(J,p,label='$p(j)$')
plt.plot(J,c,label='$c(j)$', color='r')
plt.legend()
plt.show()
```
